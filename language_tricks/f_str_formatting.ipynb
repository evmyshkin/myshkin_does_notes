# Форматрование f-строк
При печати строки можно использовать специальные операторы форматирования. Базовый синтаксис f-строки позволяет подставить значение переменной в строку.

```python
a = 'word'
print(f'Я написал слово {a}')
```

## Литералы
Можно использовать `{variable=}` чтобы в f-string подставить название переменной и ее литерал.

```python
v = 123
print(f'Переменная и ее значение: {v=}')
```

## Отступы
С помощью оператора `:`, можно указать число символов отступа справа.

```python
s = 'word'
print(f'{s:10} next word')
```

Важно, чтобы строка перед `:` была меньше, либо равна отсутпу. Иначе ничего не произойдет.

```python
s = 'long word wow' # 13 символов
print(f'{s:10} next word') # Нет смещения, отступ был бы только если строка < 10 символов
```

Можно явно указать отступ слева от строки с помощью `>`, либо справа с помощью `<`

```python
s = 'indent'
print(f'left {s:>10}') # Отступ слева
print(f'{s:<10} right') # Отступ справа
```

Обычно отступы используются для форматирования табличек

```python
print(f'{"letters":10} | {"nums":10}|')
print(f'{"abc":10} | {"123":10}|')
```

## Отсекание знаков после запятой
Используем оператор `.<число знаков>`.

```python
congrats = 'Happy birthday!'
print(f'{congrats:.5}')
```

Может использоваться совместно с отступами и другими операторами

```python
congrats = 'Happy birthday!'
name = 'Gilmore'
print(f'{congrats:<10.5}{name}')
```
