# `inspect` – исследование кода
Модуль inspect предоставляет функции для получения информации о живых объектах Python, таких как функции, классы и модули. С его помощью можно получить метаинформацию о функциях и их аргументах, исходном коде, а также узнать детали стека вызовов.

## `getsource()`
Приводит исходный код в строку для дальнейшей инспекции

```python
import inspect

def say_hi(name: str):
	"""Says hi."""
	print(f"Hi {name}!")

source = inspect.getsource(say_hi) # Преобразовать исходный код в строку
print(source)
```

## `signature()`
Проверка сигнатуры кода, преобразованного в стрку.

```python
signature = inspect.signature(say_hi) # Вывести сигнатуру, применимо только к Callable
print(signature)
```
