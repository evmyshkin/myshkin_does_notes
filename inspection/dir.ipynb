# `dir()`
Функция, которая показывает все объекты, доступные в текущем скоупе пространства имен.  Если передать объект в `dir()`, она возвращает список атрибутов и методов этого объекта. Это фактически все имена, которые можно получить через getattr() для этого объекта.

Для пользовательских объектов (классов и инстансов) результат может включать как явные атрибуты, так и методы, определенные в классе или его родителях. Для встроенных типов — их встроенные методы и атрибуты.

```python
var = 'abc'
dir(var)
```

Если аргумент не передавать вернутся все объекты пространства имен.

```python
dir(var)
```

Если вызвать `dir()` из методов экземпляра класса, вернутся объекты из пространства имен класса.

```python
from pprint import pprint

class NewClass:
	def get_dir(self):
		return dir(self)

new_obj = NewClass()
pprint(new_obj.get_dir())
```
