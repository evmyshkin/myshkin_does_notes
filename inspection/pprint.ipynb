# pprint – pretty print
Используется вместо `print()` для отладки объемных структур данных. Приводит их к человеко-читаемому виду при помощи сортировки, отступов, управления уровнями вложенности и тд.

```python
from pprint import pprint

data = {'name': 'Alice', 'age': 30, 'address': {'street': 'Main St', 'city': 'Wonderland'}}

# Обычный принт
print(data)

# Улучшенный pprint!
pprint(data)
pprint(object=data, depth=1) # Управляет глубиной вложенности и другими параметрами
```

Как и `print()` `pprint()` не рекоммендуется использовать для журналирования событий в проде. У него нет методов уровня логов, нет временных метод, форматирования под шаблоны и обработки исключений с выводом стека (как например у `logger.exception()`).
