```python
import threading
import time

a = 1

def worker1(val):
    for _ in range(5):
        val += 1
        print(f"Worker 1: {val}")
        time.sleep(1)

def worker2(val):
    for _ in range(5):
        val += 1
        print(f"Worker 2: {val}")
        time.sleep(1)

# Создаем потоки
thread1 = threading.Thread(target=worker1, args=[a])
thread2 = threading.Thread(target=worker1, args=[a])

# Запускаем потоки
thread1.start()
thread2.start()

# Ожидаем завершения потоков
thread1.join()
thread2.join()
```

## Race conditions
Когда два и более потока пытаются получить изменить или получить доступ к одним и тем же данным. Операции в потоках, которые защищают от гонки состояний называются «потокобезопасными».
Это достигается за счет вызова метода `threading.lock()` в контекстном менеджере перед выполнением основной логики взаимодействия с данными.

```python
import threading

counter = 0

def increment():
    global counter
    for _ in range(10000000):
        counter += 1

threads = [threading.Thread(target=increment) for _ in range(5)]

for t in threads:
    t.start()

for t in threads:
    t.join()

print(counter)  # Ожидаемый результат: 200000
```
