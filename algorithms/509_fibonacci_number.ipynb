# 509. Fibonacci Number

```python
class Solution:
    def fib(self, n: int) -> int:
        """
        n =0 -> 0 # Базовый сценарий: всегда 0
        n = 1 -> 1 # Базовый сценарий: всегда 1
        n = 2 -> 1 # Сумма двух предыдущим чисел: 1 + 0
        n = 3 -> 2 # Сумма двух предыдущим чисел: 1 + 1
        n = 4 -> 3 # 2 + 1
        n = 5 -> 5 # 3 + 2
        n = ...
        n = X -> X-1 + X-2 # Принцип повторяется
        """
        memo = {0: 0, 1: 1}
        if n > 1:
            for i in range(2, n + 1):
                memo[i] = memo[i - 1] + memo[i - 2]

        res = memo[n]
        return res


solution = Solution()
solution.fib(n=100)
```
