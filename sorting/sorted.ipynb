# `sorted()`
Создает отсортированную копию последовательности. Принимает аргумент `key` для критерия сортировки.

```python
l = [5, 1, 4, 11, 20, 4, 6]
sl = sorted(l)
print(sl)
```

`key` принимает тип, по значениям которого можно сортировкать. Например, если есть смешаный список из строк и цифр, можно явно указать, сортировать ли их как числа, или как строки.

```python
l = [5, 1, '4', 7, 0, '11', '20', 4, 6]
sl_int = sorted(l, key=int)
sl_str = sorted(l, key=str)
print(sl_int)
print(sl_str)
```
