# Массивы
Массивы это специализированные последовательности элементов, в которых можно тонко настраивать тип данных и управлять памятью. Типы соответствуют типам объектов в C, но в Python практически все соответствуют типу `int`.

Основные режимы создания
- 'b' — signed char (целое число)
- 'B' — unsigned char
- 'h' — signed short (целое число)
- 'H' — unsigned short
- 'i' — signed int
- 'I' — unsigned int
- 'f' — float
- 'd' — double

```python
from array import array
from sys import getsizeof
from random import randint


lst = [randint(0, 10) for n in range(10**7)]
print(f'list: {getsizeof(lst)}')
print(f'arr b:" {getsizeof(array('b', lst))}')
print(f'arr B: {getsizeof(array('B', lst))}')
print(f'arr h: {getsizeof(array('h', lst))}')
print(f'arr H: {getsizeof(array('H', lst))}')
print(f'arr i: {getsizeof(array('i', lst))}')
print(f'arr I: {getsizeof(array('I', lst))}')
```

```python
from random import random

lst_float = [random() for i in range(10**7)]
print(f'list float: {getsizeof(lst_float)}')
print(f'arr f:" {getsizeof(array('f', lst_float))}')
print(f'arr d: {getsizeof(array('d', lst_float))}')
```

```python
lst = [random() for n in range(10**7)]

print(f'Список размером {getsizeof(lst)}')

arr = array('d', (random() for i in range(10**7)))

print(f'Массив размером {getsizeof(arr)}')
```

## Создание двоичного файла из массива
Из массива создаем двоичный файл

```python
fp = open('floats.bin', 'wb') # Открыть файл для записи в двоичном формате
arr.tofile(fp)
fp.close()
```

## Чтение данных из двоичного файла для второго массива

```python
arr2 = array('d') # пустой массив с точкой двойной точности
fp = open('floats.bin', 'rb')
arr2.fromfile(fp, 10**7) # Прочитать 10 млн чисел из файла
fp.close()
arr2 == arr
```
