# memoryview
При использовании `memoryview` создается представление данных с разделением в памяти. Не создается копия.

Преобразование структуры данных с помощью `cast()` не создает копию и не меняет структуру исходного объекта. Это лишь представление.

При этом, через индекс, валидный для скасченного `memoryview`, можно менять данные в исходном объекте.

```python
from array import array

octets = array('B', range(6))

m1 = memoryview(octets)
m1.tolist()
```

```python
m2 = m1.cast('B', [2, 3])
m2.tolist()
```

```python
m3 = m1.cast('B', [3, 2])
m3.tolist()
```

```python
m2[1, 1] = 22
m2[1, 1] = 33
octets
```
