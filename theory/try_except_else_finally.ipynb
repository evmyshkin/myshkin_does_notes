# Назначение и порядок выполнения блоков try, except, else и finally в Python
В Python блоки `try`, `except`, `else` и `finally` используются для обработки исключений и выполнения завершающих действий.

1. **`try`**: В этом блоке размещается код, который может вызвать исключение.
2. **`except`**: Этот блок перехватывает исключения, возникшие в блоке `try`, и позволяет обработать ошибку.
3. **`else`**: Выполняется, если в блоке `try` не произошло исключений. Используется для кода, который должен выполняться только при отсутствии ошибок.
4. **`finally`**: Этот блок всегда выполняется, независимо от того, возникло исключение или нет. Обычно используется для очистки ресурсов, таких как закрытие файлов или освобождение памяти.

Порядок выполнения:

* Сначала выполняется `try`.
* Если исключение не произошло, выполняется блок `else` (если он присутствует).
* Если произошло исключение, выполняется соответствующий `except`.
* В любом случае выполняется блок `finally`, если он есть.


```python
def try_except_block(val: int):
	try:
		print(val + 1)
	except ValueError as exc:
		print(exc)
	else:
		print(val + 2)
	finally:
		print(3)

try_except_block(val=1)
```
