## Распаковка отображений словаря
Оператор `**` можно использовать явно для распаковки словаря, переданного одним из значений в `**kwargs`-аргумент функции.
Ключи должны быть строками и не повторяться.


```python
def dump(**kwargs):
    return kwargs

dump(**{'a': 1}, b=2, **{'c': 3})
```

А еще можно использовать `**` внутри словарного литерала. В таком случае, повторяюшиеся ключи допустимы.

```python
{'a': 1, **{'b': 2}, 'c':3, **{'d': 4, 'e': 5}}
```

## Объединение словарей
Через оператор `|`. При этом, если словари пересекаются по ключам, будет применен ключ из второго словаря.

Оператор не изменяет исходные объекты. Но можно присвоить переменным новые значения после слияения с помощью `|=`.

```python
d1 = {'a':1, 'b':2}
d2 = {'b':3, 'd': 4}
print(d1 | d2)
print(d1)
d1 |= d2
print(d1)
```
